Adott az alábbi:
$$ x_{n+1}=4x_n(1-x_n) $$
logisztikus leképezés, ahol $x$ egy 0 és 1 közötti valós szám, $n$ pedig egy természetes szám. 

 **a)** Definiáljunk egy függvényt, amely ennek a leképezésnek az első $n$ elemét számolja ki egy adott $x_0$ kezdetiértékről indulva!  
**b)** Adjuk meg az $x_{10}$ értékét a következő kezdetiértékekre: $x_0=0.199$, $x_0=0.2$, $x_0=0.201$! Mit tapasztalunk?  
**c)** Ábrázoljuk a fenti értékektől indulva az első 50 elemet egy grafikonon!

**a)** A függvény definíciója (*10 pont*):

```julia
function logistic(x₀, n)
    x = [x₀]; xₙ = x₀
    for i in 1:n
        xₙ₊₁ = 4*xₙ*(1-xₙ)
        push!(x,xₙ₊₁)
        xₙ = xₙ₊₁
    end
    return x
end
```

**b)** A tizedik elem kiszámítása három különböző kezdetiértékből kiindulva (*5 pont*):

$x_0=0.199$

```julia
x₀ = 0.199 
logistic(x₀,10)
```

$x_0=0.2$

```julia
x₀ = 0.2
logistic(x₀,10)
```

$x_0=0.201$

```julia
x₀ = 0.201
logistic(x₀,10)
```

Mit veszünk észre?

Válasz (*5 pont*): Az 5.-6. lépés esetében már divergálnak a trajektóriák, annak ellenére, hogy a kezdeti különbség csak 1%.

**c)** Ábrázolás (*10 pont*):

(*TIPP:* használjuk a `plot!()` függvényt)

```julia
#Plots.reset_defaults()
#Plots.scalefontsizes(2.5)
x₀ = 0.199 
x = logistic(x₀,50)
plot(1:51, x, label=(L"$x_{01}=0.199$"), line=(2, :red),  
              xlabel=L"n", guidefontsize=20, ylabel=L"x_n", tickfontsize=12, legendfontsize=15)

x₀ = 0.2
x = logistic(x₀,50)
plot!(1:51, x, label=(L"x_{02}=0.2"), line=(2, :black))

x₀ = 0.201 
x = logistic(x₀,50)
plot!(1:51, x, label=(L"x_{03}=0.201"), line=(2, :green))

#savefig("logistic_timeseries.svg")
```

## 1.3 Feladat (30 pont)

Adott az alábbi kezdetiérték feladat:
$$ \dot{x}=-x, \quad x(0)=1. $$

Ebben a feladatban az Euler-féle numerikus integrálási módszert fogjuk tesztelni a fenti egyenleten. Az Euler módszerról [itt](https://en.wikipedia.org/wiki/Euler_method) olvashattok többet.

Röviden a módszer a keresett függvény $t+\Delta t$ pillanatbeli értékét közelíti meg, sorbafejtve az $x(t+\Delta t)$-t a lineáris tagig:
$$ x(t+\Delta t)= x(t)+f(x,t)\Delta t,$$
ahol $\dot{x}=f(x,t)$.

**a)** Oldjuk meg a feladatot analitikusan! Mi a pontos értéke $x(1)$-nek?  
**b)** Implementáljuk az Euler módszert!   
**c)** Becsüljük meg $x(1)$-et numerikusan az Euler módszer segítségével különböző időlépésekkel (legyen ez $\hat{x}(1)$)! Használjuk a $\Delta t=\{1,0.1,10^{-2},10^{-3},10^{-4}\}$ értékeket!  
**d)** Ábrázoljuk az $E=|\hat{x}(1)-x(1)|$ abszolút hibát a $\Delta t$ időlépés függvényében! Ábrázoljuk az eredményt log-log skálán is!  

**a)** Az $x(1)$ pontos$^*$ értéke (*5 pont*):

(*TIPP*: használd a Julia beépített függvényeit)

$^*$*nyilván számítógépen számolva sosem lesz teljesen pontos, viszont vannak az Euler módszernél sokkal pontosabb módszerek is*

```julia
function x_a(t)
    return exp(-t)
end
```

```julia
x_a(1)
```

**b)** Az Euler módszer implementációja (*10 pont*):

* az egyetenlet jobboldala (RHS fügvénye), $f(x,t)$.

```julia
function f_log(x)
    return -x
end
```

* Euler módszer:  $\Delta t$ időlépés, $t$ maximális idő

```julia
t = 0.; x = [1.]
function euler!(x,t,Δt)
    n = Int(t/Δt)
    xt = x[1]
    for i in 1:n
        xt = xt + f_log(xt)*Δt
        push!(x,xt)
    end
end
```

```julia
E_dt = []
```

**c)** Az $\hat{x}(1)$ kiszámítása (*5 pont*):

$\Delta t=1$

```julia
t = 0.; x_dt0 = [1.]
euler!(x_dt0,1,1)
push!(E_dt,abs(x_dt0[end]-x_a(1)))
x_dt0[end]
```

$\Delta t=10^{-1}$

```julia
t = 0.; x_dt1 = [1.]
euler!(x_dt1,1,0.1)
push!(E_dt,abs(x_dt1[end]-x_a(1)))
x_dt1[end]
```

$\Delta t=10^{-2}$


```julia
t = 0.; x_dt2 = [1.]
euler!(x_dt2,1,0.01)
push!(E_dt,abs(x_dt2[end]-x_a(1)))
x_dt2[end]
```

$\Delta t=10^{-3}$

```julia
t = 0.; x_dt3 = [1.]
euler!(x_dt3,1,0.001)
push!(E_dt,abs(x_dt3[end]-x_a(1)))
x_dt3[end]
```

$\Delta t=10^{-4}$

```julia
t = 0.; x_dt4 = [1.]
euler!(x_dt4,1,0.0001)
push!(E_dt,abs(x_dt4[end]-x_a(1)))
x_dt4[end]
```
